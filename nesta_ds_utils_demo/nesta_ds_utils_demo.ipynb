### Upload/download from S3
Let's say we want to check what's in a bucket in S3:

````python
bucket = "nesta-ds-utils-example"
S3.get_bucket_filenames_s3(bucket)
````

We now want to upload some data on the same bucket as `csv`. We can do this using `S3.upload_obj`:

````python
help(S3.upload_obj)
````

````python
data = pd.DataFrame({"col1": [6, 1, 0, np.nan], "col2": [5, 1, 3, 5]})
S3.upload_obj(data, bucket, "data_dir/data_uploaded.parquet")
````

Now let's download the data back from S3 on memory and check it out:

````python
S3.download_obj(bucket, "data_dir/data_uploaded.csv", download_as='dataframe')
````

Oh no, the index! No problem, you can pass options as `kwargs` for the object conversion:

````python
S3.upload_obj(data, bucket, "data_dir/data_uploaded.csv", kwargs_writing={'index': False})
S3.download_obj(bucket, "data_dir/data_uploaded.csv", download_as='dataframe')
````

### Styling figures in altair

Witht the `nesta_ds_utils.viz` module we can save and styling figures on altair. 

````python
pip install vega_datasets
````

````python
import altair as alt
from vega_datasets import data

source = data.seattle_weather()
figure = alt.Chart(source).mark_bar(
         cornerRadiusTopLeft=3,
         cornerRadiusTopRight=3).encode(
         x='month(date):O',
         y='count():Q',
         color='weather:N')
figure
````

Let's apply the Nesta's theme with `setup_theme()`:

````python
from nesta_ds_utils.viz.altair import formatting, saving
formatting.setup_theme()
figure
````

````python
saving.save(figure, name='test_figure', path='Documents/figures')
````
